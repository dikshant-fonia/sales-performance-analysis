```python
import pandas as pd
df = pd.read_csv('Sample - Superstore.csv', encoding='ISO-8859-1')
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python

df.drop(['Customer Name', 'Postal Code'], axis=1, inplace=True)
```

```python
df['Profit Margin %']=df['Profit']/df['Sales']*100
```

```python
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px



```

```python
sns.set_theme(style="whitegrid")
```

```python
cat_pref = df.groupby('Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)

cat_pref.plot(kind='bar', figsize=(10,5), title='Sales & Profit by Category')
plt.ylabel("Amount ($)")
plt.show()
```

```python

df['Month-Year'] = df['Order Date'].dt.to_period('M')

monthly_sales = df.groupby('Month-Year')['Sales'].sum()

monthly_sales.plot(kind='line', figsize=(12,6), marker='o', title='Monthly Sales Trend')
plt.ylabel("Sales ($)")
plt.xlabel("Month-Year")
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='barh', figsize=(10,6), title='Top 10 Most Profitable Products', color='green')
plt.xlabel("Profit ($)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category')
plt.title("Discount vs Profit")
plt.show()
```

```python
region_perf = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Profit', ascending=False)

region_perf.plot(kind='bar', figsize=(10,5), title='Profit by Region', color=['skyblue', 'green'])
plt.ylabel("Amount ($)")
plt.show()
```

```python

seg_perf = df.groupby('Segment')[['Sales', 'Profit']].sum()
seg_perf.plot(kind='bar', title='Segment-wise Sales & Profit', figsize=(8,4))
plt.show()


ship_perf = df.groupby('Ship Mode')[['Sales', 'Profit']].sum()
ship_perf.plot(kind='bar', title='Shipping Mode Performance', figsize=(8,4))
plt.show()
```
